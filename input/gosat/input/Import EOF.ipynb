Import EOF
==========

This code is used to massage the files generated by Christian into the format needed by the static input file used by Level 2.

```ruby
import h5py
import numpy as np
```

The input data is found in /groups/algorithm/l2_fp/eof_data/gosat, we put it there because the files are really too big to go into git (they are near 1 GB in size).

We read through each of the input files, collecting "corr1" through "corr3". Unlike OCO-2, we don't have multiple footprints to work with. Note also that the file format seems to be different from what we used for OCO-2, not sure if that will continue in the future or not.

```ruby
basedir = "/groups/algorithm/l2_fp/eof_data/gosat/"
fin = h5py.File(basedir + "EOF_B4_QLERT1_June_09_2014_gainH.h5", "r")
d = []
for e in range (1, 3):
    dr = []
    for b in range(1, 4):
        dr.append(fin["/FH/corr_band%d_EOF%d" % (b, e)][:])
    d.append(dr)
```

The data hasn't been scaled yet. We scale by the standard deviation, and by 1e-8 (not sure where this comes from).

```ruby
for dr in d:
    for dv in dr:
        dv[:,1] = dv[:, 1] / np.std(dv[:, 1]) * 1e-8
```

Now write to the static input file. Note that because the size may change we delete and recreate the dataset.

If you are creating a new EOF (e.g., a test one, but leaving another one alone), the easiest thing to do to to first copy the existing EOF. That way we have unit ready already. For GOSAT, the units should be "W / cm^2 / sr / cm^-1".

```ruby
with h5py.File("l2_gosat_static_input.h5", "r+") as f:
    for e in range (1, 3):
        for b in range(1, 4):
            dname = "Instrument/EmpiricalOrthogonalFunctionTest/H_gain/EOF_%d_waveform_%d" % (e, b)
            u = f[dname].attrs["Units"]
            del f[dname]
            f[dname] = d[e - 1][b - 1]
            f[dname].attrs["Units"] = u
```

Now repeat for medium gain data.

```ruby
fin = h5py.File(basedir + "EOF_B4_QLERT1_June_09_2014_gainM.h5", "r")
d = []
for e in range (1, 3):
    dr = []
    for b in range(1, 4):
        dr.append(fin["/FM/corr_band%d_EOF%d" % (b, e)][:])
    d.append(dr)
for dr in d:
    for dv in dr:
        dv[:,1] = dv[:, 1] / np.std(dv[:, 1]) * 1e-8
with h5py.File("l2_gosat_static_input.h5", "r+") as f:
    for e in range (1, 3):
        for b in range(1, 4):
            dname = "Instrument/EmpiricalOrthogonalFunction/M_gain/EOF_%d_waveform_%d" % (e, b)
            u = f[dname].attrs["Units"]
            del f[dname]
            f[dname] = d[e - 1][b - 1]
            f[dname].attrs["Units"] = u
```
