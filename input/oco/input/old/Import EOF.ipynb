Import EOF
==========

This code is used to massage the files generated by Christian into the format needed by the static input file used by Level 2.

```ruby
import h5py
import numpy as np
```

We read through each of the input files, collecting "corr1" through "corr3". The range here is over footprint. corr1 is for the first spectral band, corr2 for the second, and corr3 for the third (i.e., these are not the first, second, and third order correction as the name might imply).

```ruby
corr1 = []
corr2 = []
corr3 = []
for i in range(1,9):
  f = h5py.File("EOF_TVAC/EOF_TVAC_FP%d.h5" % i)
  corr1.append(f["F/corr1"][:,:])
  corr2.append(f["F/corr2"][:,:])
  corr3.append(f["F/corr3"][:,:])
```

We stack the EOF together, and then transpose so this is in footprint, EOF index, wn/EOF index

```ruby
corr1 = np.transpose(np.dstack(corr1),(2,0,1))
corr2 = np.transpose(np.dstack(corr2),(2,0,1))
corr3 = np.transpose(np.dstack(corr3),(2,0,1))
```

Open the output file for writing to.

```ruby
f = h5py.File("input/oco/input/l2_oco_static_input.h5", "r+")
```

Create new output group, and populate the apriori and covariance data

```ruby
g = f["Instrument"].create_group("EmpiricalOrthogonalFunction")
a = np.zeros((3,1,1))
g.create_dataset("a_priori", data=a)
g["a_priori"].attrs["Units"] = "dimensionless"

cov = np.empty((3,1,1,1))
cov[:,:,:,:] = 100.0
g.create_dataset("covariance", data=cov)
g["covariance"].attrs["Units"] = "dimensionless"
```

Write EOF, including units.

```ruby
g.create_dataset("EOF_1_waveform_1", data=corr1)
g.create_dataset("EOF_1_waveform_2", data=corr2)
g.create_dataset("EOF_1_waveform_3", data=corr3)

g["EOF_1_waveform_1"].attrs["Units"] = "ph / s / m^2 / micron / sr"
g["EOF_1_waveform_2"].attrs["Units"] = "ph / s / m^2 / micron / sr"
g["EOF_1_waveform_3"].attrs["Units"] = "ph / s / m^2 / micron / sr"
```

```ruby
f.close()
```
